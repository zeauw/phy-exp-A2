```python
import numpy as np
import matplotlib.pyplot as plt

meaDeg = np.array([0,15,30,45,60,75,80,84,87,90])
mea = np.array([6.247,5.821,4.672,3.117,1.561,0.429,0.197,0.078,0.023,0.006])
meaNor = (mea - np.min(mea)) / (np.max(mea) - np.min(mea))
meaRad = np.pi*meaDeg / 180
err = meaNor - np.square(np.cos(meaRad))

fig1, ax = plt.subplots()
ax.scatter(meaDeg, meaNor)
xAxis = np.linspace(0,90)
yAxis = np.square(np.cos(xAxis * np.pi / 180))
ax.plot(xAxis,yAxis,'--',c='grey')
ax.set_xlim(-5,95)
ax.set_xlabel('Degree($^\circ$)')
ax.set_ylabel('normalized $I_m$')
avg = np.average(err)
print('err avg is ',avg)
plt.savefig('fig1.pdf')
```
