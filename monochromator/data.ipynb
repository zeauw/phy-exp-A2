# 曲线
电流2.21A

```python
thin = pd.read_csv('thin.csv')
thick = pd.read_csv('thick.csv')
wlThin = thin.wavelength - 0.886
wlThick = thick.wavelength - 0.886
iThin = thin.intensity
iThick = thick.intensity

fig0, ax = plt.subplots()
ax.scatter(wlThin,iThin)
ax.plot(wlThin,iThin)
ax.scatter(wlThick,iThick)
ax.plot(wlThick,iThick)
ax.set_xlabel('$\lambda (nm)$')
ax.grid()
```

```python
alpha = (np.log(iThin) - np.log(iThick)) / 0.5
fig1, ax = plt.subplots()
ax.scatter(wlThin,alpha)
ax.plot(wlThin,alpha)
ax.set_xlabel('$\lambda (nm)$')
ax.grid()
```
