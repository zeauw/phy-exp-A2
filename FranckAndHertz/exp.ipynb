# 电流随加速电压的变化

采用的加热电流为0.85A，正向电压2.5V，反向电压4V，采用的读数标准（注意，与上面不同）为等待相等的时间间隔使其相对稳定。另外，在调节电压为1V后，实际并不能达到1V，通常相差0.005V左右，直接近似认为电压为整数（其实就是偷懒）

作图，如下图


```python
data = pd.read_csv('data.csv')

fig, ax = plt.subplots()
accVolt = data['加速电压(V)']
cur = data['万用表示数(mV)']
ax.scatter(accVolt, cur)
ax.plot(accVolt,cur)
ax.set_xlabel('$V_2$/$V$')
ax.set_ylabel('$I_A$/$\mu A$')
plt.show()
```

# 对图像的计算讨论

- 激发能，采用两个极大之差所对应电压间隔之差，分别得到$12eV$，$11eV$，极小值相减，分别得到$13eV$，$11eV$，平均，得$11.75eV$

- 光的波长，采用$1e \cdot 1V=\frac{hc}{\lambda}$，有：$\lambda = 105.53 nm$

- 电子速度，忽略相对论，$v=\sqrt{\frac{2E}/{m}}=2 \cdot 10^6 m/s$
